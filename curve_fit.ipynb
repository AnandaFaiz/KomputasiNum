```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Generate linear data
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50) #Linear relationship with noise
```

```python
#Fit linear model
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y) #Reshape x for sklearn

#Predict values
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

#Plot results
plt.scatter(x, y, label='Data points')
plt.plot(xfit, yfit, color='red', label='Fitted line');
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression fit')
plt.legend()
plt.show()
```

```python
#Print model parameters
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
#Generate non-linear data
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

#Plot non-linear data
plt.scatter(x, y)
plt.title('Scatter plot of non-linear data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
